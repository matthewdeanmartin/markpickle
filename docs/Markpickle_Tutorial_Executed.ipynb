# Markpickle Tutorial

## Introduction to Markpickle

Markpickle is a Python library for **lossy serialization of Python data types to Markdown and back**. It treats Markdown as a data serialization format, allowing you to convert Python dictionaries, lists, and other simple types to readable Markdown, and parse structured Markdown back into Python objects.

### Key Features
- Serializes Python objects to human-readable Markdown
- Deserializes structured Markdown back to Python
- Supports various Markdown representations (lists, headers, tables)
- Handles binary data (images, files) with base64 encoding
- Multi-document support with horizontal line separators
- Configurable behavior for customization

### Use Cases
- Creating documentation from data
- Configuration files in Markdown format
- Data exchange between systems
- Rapid prototyping and debugging

## Basic Usage

Let's start with the most basic functionality: serializing Python objects to Markdown and deserializing Markdown back to Python.


```python
import markpickle

# Serialize Python object to Markdown
data = {"cat": {"name": "Ringo", "species": "Felix"}}
markdown = markpickle.dumps(data)
print("Serialized Markdown:")
print("=" * 50)
print(markdown)

# Deserialize Markdown back to Python object
python_obj = markpickle.loads(markdown)
print("\nDeserialized Python Object:")
print("=" * 50)
print(python_obj)
print(f"Type: {type(python_obj)}")
```

## Working with Different Data Types

Markpickle supports various data types. Let's explore them one by one.


### 1. Scalar Values

Markpickle handles strings, numbers, booleans, None, dates, and datetimes.


```python
import markpickle
from datetime import date, datetime

data = {
    "name": "John Doe",
    "age": 30,
    "is_active": True,
    "score": 98.5,
    "notes": None,
    "birthdate": date(1994, 5, 15),
    "last_login": datetime(2024, 1, 15, 10, 30, 45)
}

print(markpickle.dumps(data))
```

### 2. Lists

Lists are serialized as bullet points.


```python
import markpickle

# Simple list
data = ["apple", "banana", "orange"]
print(markpickle.dumps(data))

print("\n" + "=" * 50 + "\n")

# Nested list
data = ["fruits", ["apple", "banana"], "vegetables", ["carrot", "spinach"]]
print(markpickle.dumps(data))
```

### 3. Dictionaries

Dictionaries with scalar values are serialized as ATX headers (##, ###, etc.).


```python
import markpickle

data = {
    "person": {
        "name": "Jane Smith",
        "contact": {
            "email": "jane@example.com",
            "phone": "123-456-7890"
        },
        "hobbies": ["reading", "hiking", "photography"]
    },
    "company": {
        "name": "Tech Solutions",
        "founded": 2010
    }
}

print(markpickle.dumps(data))
```

### 4. Lists of Dictionaries

Lists of dictionaries are automatically serialized as Markdown tables.


```python
import markpickle

data = [
    {"animal": "cat", "name": "Frisky", "age": 3},
    {"animal": "dog", "name": "Fido", "age": 5},
    {"animal": "bird", "name": "Tweety", "age": 1}
]

print(markpickle.dumps(data))
```

### 5. Binary Data

Markpickle can serialize binary data (like images) as base64-encoded strings in Markdown image syntax.


```python
import markpickle

# Serialize binary data
binary_data = b"Hello, this is binary data!"
markdown = markpickle.dumps(binary_data)
print("Serialized Binary Data:")
print(markdown)

# Deserialize binary data
deserialized = markpickle.loads(markdown)
print("\nDeserialized Data:")
print(deserialized)
print(f"Type: {type(deserialized)}")
```

## Advanced Configuration

Markpickle offers configurable behavior through the `Config` class. Let's explore some customization options.


```python
import markpickle
from markpickle import Config
from datetime import date

# Create custom configuration
config = Config(
    list_bullet_style="*",  # Use asterisks instead of hyphens
    serialize_date_format="%m/%d/%Y",  # Custom date format
    true_values=["YES"],  # Recognize "YES" as True
    false_values=["NO"],  # Recognize "NO" as False
    serialize_force_final_newline=True  # Ensure file ends with newline
)

data = {
    "active": True,
    "expiration": date(2024, 12, 31),
    "tags": ["python", "markdown", "serialization"]
}

# Serialize with custom config
markdown = markpickle.dumps(data, config=config)
print("Customized Serialization:")
print(markdown)
```

## Working with Files

Markpickle provides `dump` and `load` functions to work directly with files.


```python
import markpickle
import os

data = {
    "project": "Markpickle Tutorial",
    "version": "1.0",
    "author": "John Doe",
    "tags": ["python", "markdown"]
}

# Serialize to file
with open("data.md", "w") as f:
    markpickle.dump(data, f)
print("Data serialized to data.md")

# Deserialize from file
with open("data.md", "r") as f:
    loaded_data = markpickle.load(f)
print("Data loaded from data.md:")
print(loaded_data)

# Clean up
os.remove("data.md")
```

## Multiple Documents

Markpickle supports multiple documents in a single file, separated by horizontal rules (`---`).


```python
import markpickle

# Serialize multiple documents
documents = [
    "This is the first document.",
    {"name": "Document 2", "content": "Hello world"},
    [1, 2, 3, 4]
]

markdown = markpickle.dumps_all(documents)
print("Multiple Documents:")
print(markdown)

# Deserialize multiple documents
loaded_docs = list(markpickle.loads_all(markdown))
print("\nLoaded Documents:")
for i, doc in enumerate(loaded_docs):
    print(f"Document {i+1}: {doc} ({type(doc)})")
```

## Limitations and Best Practices

### Lossy Serialization
- Markpickle is **lossy** - not all Python types or arbitrary Markdown can be perfectly serialized/deserialized
- Types like sets, tuples, and custom objects may not serialize as expected
- Precision may be lost with floating-point numbers
- Markdown formatting may affect deserialization

### Best Practices
- Use for simple data structures (dicts, lists, scalars)
- Test round-trip serialization (serialize → deserialize) for your use case
- Be cautious with complex or highly structured data
- Consider Markpickle as a human-readable serialization format, not a precise data exchange format

### Performance
- Markpickle can be slower than other serialization formats (JSON, pickle)
- For large datasets, consider other formats
- Use `dumps`/`loads` for small data, `dump`/`load` for larger files
